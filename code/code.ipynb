```python
# v√©rifier quel √©l√®ve est en √©chec global (cad il passe pas )

data['status'] = data.apply(lambda x : 'Fail' if x['math score'] == 'Fail' or 
                           x['reading score'] == 'Fail' or x['writing score'] == 'Fail'
                           else 'pass', axis = 1)

data['status'].value_counts(dropna = False).plot.bar(color = 'gray', figsize = (5, 5))
plt.title('overall results')
plt.xlabel('status')
plt.ylabel('count')
plt.show()
```

```python
# Assigning grades to the grades according to the following criteria :
# 0  - 40 marks : grade E fail 
# 41 - 60 marks : grade D moyenne 
# 60 - 70 marks : grade C normal
# 70 - 80 marks : grade B bien 
# 80 - 90 marks : grade A ca va 
# 90 - 100 marks : grade O excelent
def getgrade(percentage, status):
  if status == 'Fail':
    return 'E'
  if(percentage >= 90):
    return 'O'
  if(percentage >= 80):
    return 'A'
  if(percentage >= 70):
    return 'B'
  if(percentage >= 60):
    return 'C'
  if(percentage >= 40):
    return 'D'
  else :
    return 'E'
data['grades'] = data.apply(lambda x: getgrade(x['percentage'], x['status']), axis = 1 )

data['grades'].value_counts()
```

```python
# SVM 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns

# Charger le dataset
data = pd.read_csv(r"C:\Users\user\Desktop\python-projct\StudentsPerformance.csv")

# Afficher les premi√®res lignes pour v√©rifier le dataset
print(data.head())

# Pr√©traitement : Encoder les variables cat√©gorielles
le = LabelEncoder()
data['gender'] = le.fit_transform(data['gender'])
data['lunch'] = le.fit_transform(data['lunch'])
data['test preparation course'] = le.fit_transform(data['test preparation course'])

# Cr√©er une variable 'pass_math' indiquant si l'√©tudiant a r√©ussi en math√©matiques (score >= 50)
data['math score'] = data['math score'].apply(lambda x: 1 if x >= 50 else 0)

# S√©lection des caract√©ristiques (features) et de la cible (target)
X = data[['gender', 'reading score', 'writing score', 'lunch', 'test preparation course']]
y = data['math score']

# Diviser le dataset en ensemble d'entra√Ænement (75%) et de test (25%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Standardiser les donn√©es
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Cr√©er le mod√®le SVM
svm_model = SVC(kernel='linear', C=1)  # Utilisation d'un noyau lin√©aire

# Entra√Æner le mod√®le sur les donn√©es d'entra√Ænement
svm_model.fit(X_train, y_train)

# Pr√©dire sur les donn√©es de test
y_pred = svm_model.predict(X_test)

# √âvaluer le mod√®le
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
#Random forest 
# Importation des biblioth√®ques n√©cessaires
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Charger le dataset
data = pd.read_csv(r"C:\Users\user\Desktop\python-projct\StudentsPerformance.csv")

# Afficher les premi√®res lignes du dataset pour v√©rification
print(data.head())

# Pr√©traitement : Encoder les variables cat√©gorielles
le = LabelEncoder()
data['gender'] = le.fit_transform(data['gender'])
data['race/ethnicity'] = le.fit_transform(data['race/ethnicity'])
data['parental level of education'] = le.fit_transform(data['parental level of education'])
data['lunch'] = le.fit_transform(data['lunch'])
data['test preparation course'] = le.fit_transform(data['test preparation course'])

# Cr√©er une variable 'pass_math' indiquant si l'√©tudiant a r√©ussi en math√©matiques (score >= 50)
data['pass_math'] = data['math score'].apply(lambda x: 1 if x >= 50 else 0)

# S√©lection des caract√©ristiques (features) et de la cible (target)
X = data[['gender', 'race/ethnicity', 'parental level of education', 'lunch', 
          'test preparation course', 'reading score', 'writing score']]
y = data['pass_math']

# Diviser le dataset en ensemble d'entra√Ænement (75%) et de test (25%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Standardiser les donn√©es
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Cr√©er le mod√®le Random Forest
model = RandomForestClassifier(n_estimators=100, random_state=42)

# Entra√Æner le mod√®le sur les donn√©es d'entra√Ænement
model.fit(X_train, y_train)

# Pr√©dire sur les donn√©es de test
y_pred = model.predict(X_test)

# √âvaluer le mod√®le
print("Accuracy :", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Afficher la matrice de confusion
conf_matrix = confusion_matrix(y_test, y_pred)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Fail', 'Pass'], yticklabels=['Fail', 'Pass'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier CSV
data = pd.read_csv(r"C:\Users\user\Desktop\python-projct\StudentsPerformance.csv")

# D√©finir le seuil de r√©ussite (par exemple, 50 points)
pass_threshold = 50

# Cr√©er des colonnes indiquant si l'√©tudiant a r√©ussi chaque test
data['pass_math'] = data['math score'] >= pass_threshold
data['pass_reading'] = data['reading score'] >= pass_threshold
data['pass_writing'] = data['writing score'] >= pass_threshold

# Calculer le taux de r√©ussite pour chaque mati√®re
pass_rates = {
    'Math√©matiques': data['pass_math'].mean() * 100,
    'Lecture': data['pass_reading'].mean() * 100,
    '√âcriture': data['pass_writing'].mean() * 100
}

# Visualiser le taux de r√©ussite avec un graphique √† barres
plt.figure(figsize=(10, 6))
plt.bar(pass_rates.keys(), pass_rates.values(), color=['blue', 'orange', 'green'])
plt.title("Taux de R√©ussite des √âtudiants par Mati√®re")
plt.xlabel("Mati√®res")
plt.ylabel("Taux de R√©ussite (%)")
plt.ylim(0, 100)

# Afficher les valeurs sur chaque barre
for subject, rate in pass_rates.items():
    plt.text(subject, rate + 1, f"{rate:.2f}%", ha='center', fontsize=12)

plt.show()
```
